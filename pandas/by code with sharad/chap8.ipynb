# 3.*Changing wrong Data* ---> exceptionally high/low values, in int col float is there, etc manual errors in dataset

```python
import pandas as pd
df=pd.read_csv('dirtydata.csv')
print(df.to_string())
```

### These are called typeo errors, hence we have to tackle them manually :

like in above df 7th row has 450 and others have 45-60 range ony, hence this is a tyoeo error so we manually overrite with help of loc() [in small dataset]

```python
df.loc[7,"Duration"] = 45
print(df.to_string())
```

## making boundaries in large datasets : we use for loops

```python
df_1=pd.read_csv('dirtydata.csv')
print(df_1.to_string())

for i in df_1.index:
    if df_1.loc[i, 'Duration'] >60:
        df_1.loc[i, 'Duration'] = 60
    elif df_1.loc[i, 'Duration'] <30:
        df_1.loc[i, 'Duration'] =30
print(df_1.to_string())
```

### remove the row ---> df.drop(i, inplace=True)  ---> inside loop

```python
df_2=pd.read_csv('dirtydata.csv')
print(df_1.to_string())

for i in df_2.index:
    if df_1.loc[i, 'Duration'] >60:
        df_2.drop(i, inplace=True)
    elif df_1.loc[i, 'Duration'] <30:
        df_2.drop(i, inplace=True)

print(df_2.to_string())
```
