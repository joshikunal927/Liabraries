## 2. Cleaning wrong format :

we here too have 2 methods- 1. remove rows......... 2. convert all cells in same format

```python
import pandas as pd
df=pd.read_csv('dirtydata.csv')
print(df.to_string())
```

### 1. Method 1 ===> changing wrong datetime format :
----> df['col_name'] = pd.to_datetime(df['col_name], format="mixed")

```python
df['Date']=pd.to_datetime(df['Date'], format="mixed") # convert nan to naT and other possible datetime to datetime
print(df.to_string())
```

### 2. Method2 ===> deleting NaT rows ---> df.dropna(subset=['col_name], inplace=True)

```python
df.dropna(subset=['Date'], inplace=True)
print(df.to_string())
```
