# *Cleaning Data*
Data cleaning in pandas is the process of identifying and correcting (or removing) errors, inconsistencies, and missing values in a dataset to prepare it for analysis.


1. Empty cell 

2. Wrong format (date-time wrong format)

3. Duplicate data

4. Wrong data (exceptionally high/low value)

```python
import pandas as pd
df = pd.read_csv('dirtydata.csv')
print(df.to_string())
```

## 1. Empty Cell :
An empty cell (or missing value) in pandas refers to a cell in a DataFrame that contains no data, and is usually represented as NaN (Not a Number).

##### # Method 1. --> deleting rows with empty cells ---> df.dropna()

```python
# the below will not affect the orignal data as we are assigning a new df_variable, but if you want to change the orignal then do directly .dropna(inplace=True) ie dont store in any variable
df2= df.dropna() # drop/ delete all rows who have any column with NaN or empty value
print(df2.to_string()) 

print("\n\n\n\n\n")

# changing orignal df :=
df.dropna(inplace=True)
print(df.to_string())
```

##### # Method 2 ---> fillup with custom values ---> df.fillna(value, inplace=True)

```python
df=pd.read_csv('dirtydata.csv')
df_fill_1=df.fillna(130)
print(df_fill_1.to_string())

print("\n\n\n\n")
print(df)
```

#### 3. fillna() (specific column only with math functions or not) ----> df["col_name].fillna(value, inplace=True)

```python
df=pd.read_csv('dirtydata.csv')
mean_1=df["Calories"].mean()
print(mean_1,"\n\n")
median_1=df["Calories"].median()
print(median_1,"\n\n")
mode_1=df["Calories"].mode()[0] # mode() return a series, but for single value use [0] for first mode 
print(mode_1,"\n\n")

print(df.to_string(),"\n\n\n")


df_mean=df["Calories"].fillna(mean_1)
print(df_mean.to_string(),"\n\n\n\n")


df_median=df["Calories"].fillna(median_1)
print(df_median.to_string(),"\n\n\n\n")


df_mode=df["Calories"].fillna(mode_1)
print(df_mode.to_string(),"\n\n\n\n")
```
