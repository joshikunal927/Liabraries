```python
import matplotlib.pyplot as plt
```

# Axis 

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)
plt.show()
```

## (i) xticks and yticks ----> to change the points names

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)
plt.xticks(x,labels=["kunal","joshi","kunjos","divine","joskun","kunjos"])
plt.yticks(y,labels=["kunal","joshi","kunjos","divine","joskun","kunjos"])
plt.show()
```

## (ii) xlim() and ylim() to increase the limits of x and y axis

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,30)
plt.show()
```

## (iii) limits by axis==> axis(x_start, x_end, y_start, y_end)

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)

plt.axis([0,20,0,16])
plt.show()
```
