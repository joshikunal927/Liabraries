```python
import matplotlib.pyplot as plt
```

## (i) Basic - labels, title

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y)

plt.title("My Workings",fontsize=15)
plt.show()
```

## (ii) Explode, colors

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]

ex=[0.1,0.1,0.3,0.1]
c=["r","m","y","skyblue"]
plt.pie(x,labels=y,explode=ex,colors=c)

plt.title("My Workings",fontsize=15)
plt.show()
```

## (iii) Show percentage (autopct="%0.1f%%")

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y, autopct="%0.2f%%")  #here 0.1 means one decimal place, 0.2 means two decimal places and so on

plt.title("My Workings",fontsize=15)
plt.show()
```

### (iv) shadow, increasing size (radius), labeldistance

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y, shadow=True, radius=2, labeldistance=1.2)

plt.title("My Workings",fontsize=15)
plt.show()
```

### (v) Starting angle (default by 0 deg.)

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y, startangle=130)

plt.title("My Workings",fontsize=15)
plt.show()
```

### (vi) For changing text properties in the pie chart use ---> textprops={"parameter":value , "parameter2":value2, etc} (in dict form)

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y,autopct="%0.1f%%",radius=1.5, textprops={"fontsize":20, "color":"m" })

plt.title("My Workings",fontsize=15)
plt.show()
```

### (vii) Direction - clockwise or anticlockwise(default):

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y , counterclock=False)

plt.title("My Workings",fontsize=15)
plt.show()
```

### (viii) Edge property change- wedgeprops={"para1":value1, "para2":value2, soo onn} (in form of dict)

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y, shadow=True,wedgeprops={"linewidth":5, "width":0.8, "edgecolor":"pink"})

plt.title("My Workings",fontsize=15)
plt.show()
```

### (ix) Changing center - center=(x,y)

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y, center=(-9,9))

plt.title("My Workings",fontsize=15)
plt.show()
```

### (x) Rotating label , legend

```python
x=[30,40,10,20]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y, rotatelabels=True)

plt.title("My Workings",fontsize=15)
plt.legend(loc="lower left")
plt.show()
```

### (xi) Making dot pie chart:

```python
plt.pie([1]) # 1 is radius here
plt.show()
```

### (xii) Donut pie chrt, multiple pie chart

```python
x=[30,40,10,20]

z=[33,13,50,4]
y=["DSA","DataAnalysis","Linkdin","github"]
plt.pie(x,labels=y, radius=2 )



c=["m","skyblue","pink","b"]
plt.pie(z, radius=1.5,colors=c)

plt.pie([0.5], colors="w") # donut
plt.show()
```
