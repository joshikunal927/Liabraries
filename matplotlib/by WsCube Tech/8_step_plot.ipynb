###  Marker
(i) apply marker

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,4,2,6,1,3,4]

plt.step(x,y, marker="o")
plt.show()
```

(ii) marker size 

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,4,2,6,1,3,4]

plt.step(x,y, marker="o", ms=20)
plt.show()
```

(iii) marker face color

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,2,3,4,2,6,1,3,4]

plt.step(x,y, marker="o", mfc="skyblue")
plt.show()
```
