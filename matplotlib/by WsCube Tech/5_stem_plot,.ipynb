```python
import matplotlib.pyplot as plt
```

# Stem Plot = (dot+linear plot)

x=[ ]

```python
x= [2,5,3,9,7,8]
y=[2,-7,4,9,1,-2]

plt.stem(x,y)
plt.show()
```

### (i) changing line style ---> linefmt

linefmt	Effect

'r-'	Red solid line

'g--'	Green dashed line

'b:'	Blue dotted line

'k-.'	Black dash-dot line

```python
x= [2,5,3,9,7,8]
y=[2,-7,4,9,1,-2]

plt.stem(x,y, linefmt="r:")
plt.show()
```

### (ii) changing marker ---> markerfmt

```python
x= [2,5,3,9,7,8]
y=[2,-7,4,9,1,-2]

plt.stem(x,y, markerfmt="r+")
plt.show()
```

### (iii) changing bottom, basefmt

```python
x= [2,5,3,9,7,8]
y=[2,-7,4,9,1,-2]

plt.stem(x,y, bottom=4, basefmt="pink")
plt.show()
```

### (iv) Changing orientation

```python
x= [2,5,3,9,7,8]
y=[2,-7,4,9,1,-2]

plt.stem(x,y, orientation='horizontal')
plt.show()
```
