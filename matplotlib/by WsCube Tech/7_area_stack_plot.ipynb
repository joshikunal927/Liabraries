```python
x=[1,2,3,4,5,6,7]
y=[1,4,1,4,4,3,1]

plt.stackplot(x,y)
plt.show()
```

### (i) multiple areas :

x: Your x-axis values (shared by all)

y, y1, y2: These are stacked on top of each other.

First area = y

Second area = y + y1 (stacked on top)

Third area = y + y1 + y2 (stacked higher)



```python
x=[1,2,3,4,5,6,7]
y=[1,4,1,4,4,3,1]
y1=[2,5,1,8,4,2,7]
y2=[4,3,2,1,6,7,8]


plt.stackplot(x,y,y1,y2)
plt.show()
```

### (ii) labels, colors

```python
x=[1,2,3,4,5,6,7]
y=[1,4,1,4,4,3,1]
y1=[2,5,1,8,4,2,7]
y2=[4,3,2,1,6,7,8]


plt.stackplot(x,y,y1,y2, labels=["area1","area2","area3"] , colors=["m","skyblue","pink"])

plt.legend()
plt.show()
```

### (iii) changing baseline

 Common baseline Options:

Baseline	---> What it Does	--->Looks Like

'zero'	---> Default. Starts stacking from Y = 0	---> 📊 Standard stacked chart

'sym'	---> Symmetrical stacking around a central axis	---> 🔁 Centered around 0

'wiggle'	---> Minimizes wiggle in layers — good for stream graphs	---> 🌊 Smoothed layers

'weighted_wiggle'	---> Like wiggle, but weighted by data magnitude	---> 🎯 Smoothed + magnitude-based

```python
x=[1,2,3,4,5,6,7]
y=[1,4,1,4,4,3,1]
y1=[2,5,1,8,4,2,7]
y2=[4,3,2,1,6,7,8]


plt.stackplot(x,y,y1,y2, baseline="zero")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,4,1,4,4,3,1]
y1=[2,5,1,8,4,2,7]
y2=[4,3,2,1,6,7,8]


plt.stackplot(x,y,y1,y2, baseline="sym")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,4,1,4,4,3,1]
y1=[2,5,1,8,4,2,7]
y2=[4,3,2,1,6,7,8]


plt.stackplot(x,y,y1,y2, baseline="wiggle")
plt.show()
```
