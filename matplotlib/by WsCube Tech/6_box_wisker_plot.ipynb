# BOX WISKER PLOT

A box plot shows the distribution of a dataset using five summary statistics:

Minimum

Q1 (25th percentile)

Median (Q2, 50th percentile)

Q3 (75th percentile)

Maximum

It also shows:
Interquartile Range (IQR) = Q3 - Q1

Outliers (data points that are far from the rest)

ðŸ“¦ How It Looks:
Box â†’ from Q1 to Q3 (middle 50% of data)

Line inside box â†’ the median (Q2)

Whiskers â†’ from box to min & max (excluding outliers)

Dots outside whiskers â†’ outliers



### Syntax:

x=[ ]

plt.boxplot(x)

plt.show()

```python
x=[22,33,44,55,66,77,88,99,155]
plt.boxplot(x)
plt.show()
```

### (i) shape (notch), orientation (vertical or horizontal), width

```python
x=[22,33,44,55,66,77,88,99]
plt.boxplot(x, notch=True, vert=False, widths=0.5)
plt.show()
```

### (ii) Label, fill color box 

```python
x=[22,33,44,55,66,77,88,99,155]
plt.boxplot(x, tick_labels=["kunal"], patch_artist=True)

plt.show()
```

### (iii) show mean, whis (to join with point)

```python
x=[22,33,44,55,66,77,88,99,155]
plt.boxplot(x, showmeans=True, whis=3)
plt.show()
```

### (iv) change outler

```python
x=[22,33,44,55,66,77,88,99,155]
plt.boxplot(x, sym="r+")
plt.show()
```

### (v) box properties :

```python
x=[22,33,44,55,66,77,88,99,155]
plt.boxplot(x, boxprops={"color":"skyblue"})
plt.show()
```

### (vi) capprops :

```python
x=[22,33,44,55,66,77,88,99,155]
plt.boxplot(x, capprops={"color":"y"})
plt.show()
```

### (vii) whiskerprops :

```python
x=[22,33,44,55,66,77,88,99,155]
plt.boxplot(x, whiskerprops={"color":"m"})
plt.show()
```

### (viii) Multiple box plots :

```python
x=[22,33,44,55,66,77,88,99,155]
y=[10,20,30,40,50,60,70,80,90]

z=[x,y]

plt.boxplot(z, tick_labels=["first","second"])
plt.show()
```
