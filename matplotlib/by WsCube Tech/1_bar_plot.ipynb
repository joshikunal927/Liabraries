# 1. Bar plot :
A bar graph (or bar chart) represents categorical data with rectangular bars. Each bar's height shows the value or frequency of a specific category.

Used for: Comparing different categories (like departments, cities, product sales, etc.)

x=[]

```python
#making a simple bar graph
x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]
plt.bar(x,y)
plt.show()                                                                      
```

(iii) changing width of bars : plt.bar(x,y, width=0_to_1)

```python
plt.xlabel("student names", fontsize=15) #we can also change the font size of labels and title by using fontsize=
plt.ylabel("attendence percentage",fontsize=20)
plt.title("student vs attendence",fontsize=30)

x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]

colorz=np.array(["g","skyblue",0.66,'#33FFCE',0.1,'#ff9999'])  # color array
plt.bar(x,y, width=0.3, color=colorz) #color and width ajdust
plt.show()
```

(v) taking x parameters on edge of bar not center : ---> plt.bar(x,y, align="edge")

```python
x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]
plt.bar(x,y, align="edge") #edge
plt.show()

x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]
plt.bar(x,y, align="center") #center
plt.show()
```

(vi) changin barz edge color and thickness and line style ---> plt.bar(x,y, edgecolor="edge_color_list_or_name", linewidth=value, linestyle="style")

```python
x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]

colorz=np.array(["g","skyblue",0.66,'#33FFCE',0.1,'#ff9999'])
plt.bar(x,y,width=0.3, align="edge", edgecolor=colorz, linewidth=5, linestyle="--")
plt.show()
```

## # Making multiple bar graphs :

1. overlapping - not preffered

```python
x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]
z=[70,10,20,50,80,40]

width=0.3
plt.bar(x,y,width,color="r", label="attendence sem 1") 
plt.bar(x,z,width,color="y", label="attendence sem 2") 
plt.legend()
plt.show()
```

2. making multiple bars sideways:

```python
x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]
z=[70,10,20,50,80,40]

width=0.3
p=np.arange(len(x))
p1=[i+width for i in p]


plt.bar(p,y,width,color="r", label="attendence sem 1") 
plt.bar(p1,z,width,color="y", label="attendence sem 2") 

plt.xticks(p+width/2, x, rotation=45 )
plt.legend(loc="upper left")
plt.show()
```

3. making horizontl bar graph ---> plt.barh()

```python
x=["kunal","lokesh","kunwar","pooja","karan","kiran"]
y=[65,80,90,90,60,75]
z=[70,10,20,50,80,40]

width=0.3
p=np.arange(len(x))
p1=[i+width for i in p]


plt.barh(p,y,width,color="r", label="attendence sem 1") 
plt.barh(p1,z,width,color="y", label="attendence sem 2") 

plt.yticks(p+width/2, x, rotation=45 )
plt.legend(loc="lower right")

plt.ylabel("names")
plt.xlabel("percentage")

plt.show()
```
