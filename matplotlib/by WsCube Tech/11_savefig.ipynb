```python
import matplotlib.pyplot as plt 
```

# Savefig

### plt.savefig(figname)




.png (best for general use)

.jpg or .jpeg

.pdf (for vector format, printing)

.svg (good for web)

.eps (for publications)



```python
x=[1,2,3,4,5]
y=[2,8,2,1,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="skyblue")

x_pie=[10,20,30,40]
plt.subplot(2,2,2)
plt.pie(x_pie)

x_bar=[1,2,3,4,5]
y_bar=[10,20,15,40,30]
plt.subplot(2,2,3)
plt.bar(x_bar,y_bar)

plt.subplot(2,2,4)
plt.scatter(x,y)

plt.savefig("try.png")
plt.show()
```

## (i) For better quality ----> dpi

```python
x=[1,2,3,4,5]
y=[2,8,2,1,4]
plt.subplot(2,2,1)
plt.plot(x,y,color="skyblue")

x_pie=[10,20,30,40]
plt.subplot(2,2,2)
plt.pie(x_pie)

x_bar=[1,2,3,4,5]
y_bar=[10,20,15,40,30]
plt.subplot(2,2,3)
plt.bar(x_bar,y_bar)

plt.subplot(2,2,4)
plt.scatter(x,y)

plt.savefig("try2.png", dpi=4000)
plt.show()
```
