```python
x=[1,2,3,4,5,6,7]
y=[1,8,2,4,6,8,2]

plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,8,2,4,6,8,2]

plt.plot(x,y, color="pink")
plt.fill_between(x=[1,4],y1=0,y2=9, color="skyblue")
plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([1,8,2,4,6,8,2])

plt.plot(x,y, color="pink")
plt.fill_between(x,y, color="skyblue", where=(x>2) & (x<7) & (y>2) )
plt.show()
```
