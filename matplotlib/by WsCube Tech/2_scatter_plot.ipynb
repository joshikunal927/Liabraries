```python
x=np.array([3,5,7,11,4,3,2,9])
y=np.array([20,30,10,50,20,88,10,60])
plt.scatter(x,y, alpha=0.9, s=[90,80,70,10,20,10,90,2], linewidth=4, c=["g","r","skyblue","m","r","y","b","b"], edgecolors="m")



plt.title("this is scatter plot", fontsize=15)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

### Marker:
Marker Symbol	Description

'o'	Circle

's'	Square

'^'	Triangle Up

'v'	Triangle Down

'>'	Triangle Right

'<'	Triangle Left

'*'	Star

'D'	Diamond

'p'	Pentagon

'x'	X

'+'	Plus

'.'	Pixel point

'_'	Horizontal line

```python
x=np.array([3,5,7,11,4,3,2,9])
y=np.array([20,30,10,50,20,88,10,60])
plt.scatter(x,y, alpha=0.5, marker="^")



plt.title("this is scatter plot", fontsize=15)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

### cmap color codes :
1 to 100



```python
x=np.array([3,5,7,11,4,3,2,9])
y=np.array([20,30,10,50,20,88,10,60])
colorzz=[1,20,43,67,89,100,39,75]
plt.scatter(x,y, alpha=0.9,cmap="OrRd",c=colorzz)

plt.colorbar().set_label("this is color pallette",fontsize=25)



plt.title("this is scatter plot", fontsize=15)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

## More than one scatter_plot graph:

```python
x=np.array([3,5,7,11,4,3,2,9])
y=np.array([20,30,10,50,20,88,10,60])
z=np.array([40,100,40,29,55,19,97,30])
colorzz=[1,20,43,67,89,100,39,75]
plt.scatter(x,z,c="r")
plt.scatter(x,y, alpha=0.9,cmap="OrRd",c=colorzz)

plt.colorbar().set_label("this is color pallette",fontsize=25)



plt.title("this is scatter plot", fontsize=15)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```
