```python
import matplotlib.pyplot as plt
```

# Text In Matplotlib

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)
plt.show()
```

## (i) title, xlabel, ylabel 

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)
plt.title("NotebooK", fontsize=20)
plt.xlabel("right page")
plt.ylabel("left page")
plt.show()
```

## (ii) Text ---> plt.text(x_coordinate, y_coordinate, "text", other_parameters)

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)

plt.text(3,7,"pointer", fontsize=13, style="italic")
 
plt.show()
```

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)

plt.text(3,7,"pointer", fontsize=13, style="italic", bbox=dict( facecolor="skyblue", ))

plt.show()
```

### (ii) Annotate ---> for pointing

```python
x=[1,3,4,5,6,7]
y=[4,3,5,2,8,6]

plt.plot(x,y)
 
 
plt.annotate("pointer2",xy=(3,3),xytext=(2,5),arrowprops=dict(facecolor="pink", shrink=100))
plt.show()
```
